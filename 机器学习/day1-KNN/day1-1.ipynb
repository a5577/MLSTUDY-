#数据采集 数据处理 特征工程（特征提取、特征预处理、特征降维、特征的选取、组合） 模型训练 模型评估/预测
#模型拟合问题：拟合模型在训练集和测试集的表现情况   欠拟合00模型过于简单 过拟合10模型过于复杂有脏数据训练集太少 正好拟合11
#泛化：模型在测试集上表现好坏的能力  多个模型泛化能力相差不大 选择模型更简单的 （奥卡姆剃刀原则）
#欠拟合可以增加特征维度  过拟合可以降低维度

#KNN算法既可以分类也可以回归  k值过小会导致过拟合 过大会导致欠拟合会导致按照最多的类别进行预测
#找到合适的K值 交叉验证 网格搜素
#分类问题和回归问题（投票）：都为有监督学习
#回归是将计算出的距离的最近的k个样本进行取均值作为预测值

```python
#导包
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
#分类器 = 投票
x_train = [[0],[1],[2],[3]]#训练集数据  有多个 所以要有二维
y_train = [0,0,1,1]
x_test= [[5]]#和测试集一样
#创建模型对象
estimator = KNeighborsClassifier(n_neighbors=2)#默认是欧氏距离
#模型训练 传入训练集数据
estimator.fit(x_train,y_train)
#模型预测
y_pred = estimator.predict(x_test)
print(y_pred)
```

```python
#回归问题
x_train = [[0,0,1],[1,1,0],[3,10,10],[4,11,12]]
y_train = [0.1,0.2,0.3,0.4]
x_test = [[1,1,5]] 
#创建模型对象
estimator = KNeighborsRegressor(n_neighbors=2)
#模型训练
estimator.fit(x_train,y_train)
#模型预测
y_pred = estimator.predict(x_test)
print(y_pred)
```

```python
#数据归一化：
from sklearn.preprocessing import MinMaxScaler#归一化对象
#数据集 归一化之前
x_train = [[90,6,10,40],[60,4,15,45],[75,3,13,46]]
#创建归一化对象
transfer = MinMaxScaler(feature_range=(5,7))#默认为【0，1】
#队员数据集进行归一化
x_train_new = transfer.fit_transform(x_train)
print(x_train_new)
```

```python
from sklearn.preprocessing import StandardScaler
x_train = [[90,6,10,40],[60,4,15,45],[75,3,13,46]]
#创建标准化对象
transfer = StandardScaler()#默认为【0，1】
#队员数据集进行标准化
x_train_new = transfer.fit_transform(x_train)
print(x_train_new)
#打印方差和均值
print(transfer.mean_)
print(transfer.scale_)#方差
```
