```python
#朴素贝叶斯情感分类
import pandas as pd
import numpy as np
import jieba
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer#词频统计包 把评论内容转成词频矩阵
from sklearn.naive_bayes import MultinomialNB#贝叶斯
from sklearn.metrics import accuracy_score
import joblib
```

```python
#读取文件
data = pd.read_csv('./书籍评价.csv',encoding = 'gbk')
#data.info()
#数据预处理
#添加一列labels
data['labels'] = np.where(data['评价']=='好评',1,0)
#print(data)
#抽取labels列作为标签
y = data['labels']
#jieba分词
#数据格式[[],[],[]]→列表套字符串
comment_list = [','.join(jieba.lcut(line))for line in data['内容']]
#print(comment_list)
#加载停用词列表
with open('./stopwords.txt','r',encoding='utf-8') as f:
    #依次读取所有数据
    stopwords_list = f.readlines()
    stopwords_list =[line.strip() for line in stopwords_list]
    #去重
    stopwords_list = list(set(stopwords_list))
#删除提用词 创建向量化对象
transfer = CountVectorizer(stop_words=stopwords_list)
#统计词频矩阵
x = transfer.fit(comment_list)
#print(x)
x = transfer.transform(comment_list).toarray()
#print(x)
#获取37个好词有用词
#print(transfer.get_feature_names_out())
#数据较少 10条当训练集 后面当测试集
x_train = x[:10]
x_test = x[10:]
y_train = y[:10]
y_test = y[10:]
#模型训练
estimator = MultinomialNB()
estimator.fit(x_train,y_train)
#模型预测
y_pred = estimator.predict(x_test)
print(y_pred)
#模型评估
print(accuracy_score(y_test,y_pred))
```
