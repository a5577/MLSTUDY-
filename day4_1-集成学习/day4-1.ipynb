```python
#泰坦尼克号
#线性回归和决策树回归的对比
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier      # 回归决策树
#from sklearn.linear_model import LinearRegression   # 线性回归
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
#数据加载
data = pd.read_csv('./titanic_train.csv')
#data.info()
#数据预处理
#提取特征
x = data[['Pclass', 'Sex', 'Age']]
y = data['Survived']
#数据填充 age列有缺失 我们用该列的平均值进行填充
#数据复制 进行填充
x = x.copy()
x['Age'] = x['Age'].fillna(x['Age'].mean())
#x.info()
#将性别变成onehot
x = pd.get_dummies(x,columns=['Sex'])
x.info()
#划分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23)
#特征工程
#模型训练
#决策树
estimator1 = DecisionTreeRegressor()
estimator1.fit(x_train,y_train)
y_pred1 = estimator1.predict(x_test)
print(estimator1.score(x_test,y_test))
print('-'*23)
#随机森林
estimator2 = RandomForestClassifier(n_estimators=100,max_depth=None)
#模型训练
estimator2.fit(x_train,y_train)
y_pred2 = estimator2.predict(x_test)
print(estimator2.score(x_test,y_test))
print('-'*23)
#随机森林网格搜索
#estimator3 = RandomForestClassifier(n_estimators=100,max_depth=None)
params = {"n_estimators":[60,90,110,130,150], "max_depth": [2,3,5,7]}
#网格交叉验证
gs_estimator = GridSearchCV(estimator2,param_grid=params,cv =2) #模型 网格搜素的参数 几折
gs_estimator.fit(x_train,y_train)
y_pred3 = gs_estimator.predict(x_test)
print(gs_estimator.score(x_test,y_test))
print(gs_estimator.best_params_)
```
