```python
#AdaBoost自适应提升
#葡萄酒分类
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier      # 回归决策树
#from sklearn.linear_model import LinearRegression   # 线性回归
from sklearn.ensemble import AdaBoostClassifier#分类器
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder#标签编码器
from sklearn.metrics import accuracy_score
```

```python
#数据读取
data = pd.read_csv('./wine0501.csv')
data.info()
#print(data['Class label'].unique())
#数据预处理
#从标签列过滤掉两种类别
data = data[data['Class label'] != 1]
#print(data['Class label'].unique())
#获取特征列
x = data[['Alcohol', 'Hue']]
y = data['Class label']
#标签编码器转化为01
le = LabelEncoder()
y = le.fit_transform(y)
#print(y)
#数据划分
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23)
#创建模型单一决策树 弱分类器
estimator1 = DecisionTreeClassifier(max_depth=3)
#模型训练
estimator1.fit(x_train,y_train)
#预测
y_pred1 = estimator1.predict(x_test)
#模型评估
print(accuracy_score(y_test,y_pred1))
#集成学习Ada CART树 200颗
estimator2 = AdaBoostClassifier(estimator = estimator1,n_estimators=500,learning_rate=0.01)#决策树对象  决策树的个数 学习率
#训练模型
estimator2.fit(x_train,y_train)
#预测
y_pred2 = estimator2.predict(x_test)
#模型评估
print(accuracy_score(y_test,y_pred1))
```
