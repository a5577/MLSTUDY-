```python
import numpy as np
import pandas as pd
```

```python
#数据预处理  时间格式化按照时间升序排列且对数据去重  
```

```python
def data_preprocessing():
    data = pd.read_csv('../data/train.csv')
    #data.info()
    #时间格式化 %Y%m%d%H%M%S
    data['time'] = pd.to_datetime(data['time']).dt.strftime('%Y-%m-%d %H:%M:%S')
    #print(data.head())
    #按照时间升序排列
    data.sort_values('time',ascending=True,inplace = True)
    #print(data)
    return data
```
